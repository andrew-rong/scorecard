```python
# Load Excel file (skip header rows used for titles)
df = pd.read_excel("historical-cpi-u-202602.xlsx", skiprows=4)

# Rename columns
df.columns = [
    "Indent", "Year", "Jan", "Feb", "Mar", "Apr",
    "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

# Remove unnecessary column
df = df.drop(columns=["Indent"])

df_long = df.melt(
    id_vars="Year",
    var_name="Month",
    value_name="CPI"
)

# Optional: create a proper datetime column
# df_long["Date"] = pd.to_datetime(df_long["Year"].astype(str) + "-" + df_long["Month"])
# Convert from wide to long format


# Sort chronologically
# df_long = df_long.sort_values("Date")

print(df_long.head())

# Save to new file
df_long.to_csv("cpi_monthly_long_format.csv", index=False)
```

```python
df = pd.read_excel('historical-cpi-u-202602.xlsx', header=None)
data = df.iloc[5:119].copy()
data.columns = ['indent', 'Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data = data[['Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].copy()
data['Year'] = data['Year'].astype(int)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_order = {m: i+1 for i, m in enumerate(months)}

melted = data.melt(id_vars='Year', value_vars=months, var_name='Month', value_name='CPI')
melted = melted.sort_values(['Year']).reset_index(drop=True)
melted['CPI'] = pd.to_numeric(melted['CPI'].replace({'–': None, '-': None}), errors='coerce')
melted = melted.dropna(subset=['CPI'])
melted['Month-Year'] = melted['Month'] + '-' + melted['Year'].astype(str)
melted = melted[['Month-Year', 'CPI']].reset_index(drop=True)
melted.columns = ['Month-Year', 'CPI']

melted.to_excel('historical-cpi-u-reorganized.xlsx', index=False)
```
